```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

DATA_PATH = "API_SP.POP.TOTL_DS2_en_csv_v2_321661.csv"
ID_COLS = ["Country Name", "Country Code", "Indicator Name", "Indicator Code"]
EXCLUDE_CODES = ["WLD", "HIC", "MIC", "LIC", "LMC", "UMC"]
```

```python
# Cell 2: Load and prepare data
df_raw = pd.read_csv(DATA_PATH, skiprows=4)  # World Bank WDI export [file:22]
year_cols = [c for c in df_raw.columns if c not in ID_COLS]
last_year = max(year_cols)

df = df_raw[ID_COLS + [last_year]].copy()
df.rename(columns={last_year: "Population"}, inplace=True)

# drop aggregates and missing population
df = df[~df["Country Code"].isin(EXCLUDE_CODES)]
df = df.dropna(subset=["Population"])

print(f"Latest year in dataset: {last_year}")
df.head()
```
